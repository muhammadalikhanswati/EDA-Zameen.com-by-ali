```python
# Loading data

filename = "/kaggle/input/zameencom-property-data-pakistan/Property_with_Feature_Engineering.csv"
# Making the Data Global
global data
data = pd.read_csv(filename, sep = ',')
data.head(5)
```

```python
#1. How many records are in the dataset?
print("Number of records:", len(data))

#2. What columns are present in the dataset?

print("Columns:", data.columns.tolist())
```

```python
# 1. Get the count of each property type
property_type_counts = data['property_type'].value_counts()

# 2. Display the top N most common property types (e.g., top 10)
top_property_types = property_type_counts.head(10)
print("Top 10 Most Common Property Types:")
print(top_property_types)
```

```python
# Count how many listings exist for each purpose
print(data['purpose'].value_counts())

# Show percentages instead of counts
print(data['purpose'].value_counts(normalize=True).round(2))
```

```python
avg_price = data['price'].mean()
print(f"Average price: {avg_price:,.0f}")  # Formats with commas


min_price = data['price'].min()
max_price = data['price'].max()


print(f"Price range: {min_price:,.0f} to {max_price:,.0f}")

```

```python
# Top 10 cities by number of listings
top_cities = data['city'].value_counts().head(10)
print(top_cities)
```

```python
price_stats = data['price'].describe(percentiles=[.25, .5, .75])
print(price_stats.apply(lambda x: f"{x:,.0f}"))
```
